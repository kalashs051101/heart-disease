# 1.How many people have heart disease and how many have not heart diseases

```python
df.target.value_counts()
```

```python
df.target.value_counts().plot(kind='bar',color = ['orchid','salmon'])  # plot the bar graph
plt.title("Heart Disease")
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
```

```python
df.target.value_counts().plot(kind='pie')   # plotting the pie chart
plt.legend(['disease','no diseases'])
```

# 2.People of which sex has more heart disease

```python
df.sex.value_counts()
```

```python
df.sex.value_counts().plot(kind = 'pie')
plt.legend(['male','female'])
```

```python
pd.crosstab(df.sex,df.target)
```

```python
sns.countplot(x = 'target' , data = df ,hue = 'sex')
```

# 3.people of which sex has which type of  chest pain most?

```python
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind = 'bar')
plt.legend(['heart vs count'])
```

```python
pd.crosstab(df.sex,df.cp)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind='bar')
plt.title("type od chest pain for sex")
plt.xlabel("0=female , 1= male")
```

# 5.age vs maximum heart rate when people have heart disease
# 6.age vs maximum heart rate when people doesn't have heart disease

```python
sns.displot(x= 'age',data = df ,bins = 30, kde= True)
```

```python
sns.kdeplot(x= 'age' , y ='thalach' , data= df, color= 'darkgreen')
```

```python
plt.figure(figsize= (10,6))
plt.scatter(df.age[df.target==1],
           df.thalach[df.target==1])
plt.title("max heart rate vs age")
plt.xlabel("age")
plt.ylabel("max heart rate")
```

```python
plt.figure(figsize= (10,6))
plt.scatter(df.age[df.target==1],
           df.thalach[df.target==1])
plt.scatter(df.age[df.target==0],
           df.thalach[df.target==0])

plt.title("Max Heart Rate vs Age")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.legend(['disease','no disease'])
```
